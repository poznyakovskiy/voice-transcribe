# Audio Transcription
Batch-transcribe audio files using Vosk.

```python
# --- Configuration ---
from os.path import expanduser

input_folder  = expanduser('/data/GDrive/Audio Notes')
output_folder = expanduser('~/Documents/Transcribed Notes')
extensions    = ['.m4a', '.mp3', '.ogg', '.wav']
```

```python
# --- Discover audio files ---
import os
import glob

audio_files = [
    f
    for ext in extensions
    for f in sorted(glob.glob(os.path.join(input_folder, f'*{ext}')))
]

print(f"Found {len(audio_files)} file(s):")
for f in audio_files:
    print(' ', f)
```

```python
# --- Transcribe ---
import sys
sys.path.insert(0, os.path.dirname(os.path.abspath('transcribe.py')))

from transcribe import transcribe_file

os.makedirs(output_folder, exist_ok=True)

for audio_path in audio_files:
    base_name = os.path.splitext(os.path.basename(audio_path))[0]
    txt_path  = os.path.join(output_folder, f'{base_name}.txt')

    if os.path.exists(txt_path):
        print(f'[skip]  {base_name}')
        continue

    print(f'[proc]  {audio_path} ...', end=' ', flush=True)
    try:
        text = transcribe_file(audio_path, output_folder)
        preview = text[:80].replace('\n', ' ')
        print(f'OK — "{preview}"')
    except Exception as e:
        print(f'ERROR — {e}')
```
